## Exercise 2

```python
def camera_b_distance(f, g):
    """
    camera_b_distance returns the distance (b) where the CCD should be placed
    when the object distance (g) and the focal length (f) are given
    :param f: Focal length (in mm)
    :param g: Object distance (in m)
    :return: b, the distance where the CCD should be placed (in mm)
    """

    # Conversion to same units
    g_mm = g*1000

    if g_mm== f: # Denominator would become 0
        raise ValueError("Object at focal distance → image at infinity.")

    return (1/f - 1/g_mm) ** (-1) 
```

```python
# ALTERNATIVE

def camera_b_distance2(f, g):
    """
    camera_b_distance returns the distance (b) where the CCD should be placed
    when the object distance (g) and the focal length (f) are given
    :param f: Focal length (in mm)
    :param g: Object distance (in m)
    :return: b, the distance where the CCD should be placed (in mm)
    """
    
    # Conversion to same units
    g_mm = g*1000

    if g_mm == f: # Denominator would become 0
        raise ValueError("Object at focal distance → image at infinity.")
    
    return (f * g_mm) / (g_mm - f)
```

```python
print('f=15mm, g=0.1m:', camera_b_distance(15, 0.1))
print('f=15mm, g=1m:', camera_b_distance(15, 1))
print('f=15mm, g=5m:', camera_b_distance(15, 5))
print('f=15mm, g=15m:', camera_b_distance(15, 15))
```

```python
print('f=15mm, g=0.1m:', camera_b_distance2(15, 0.1))
print('f=15mm, g=1m:', camera_b_distance2(15, 1))
print('f=15mm, g=5m:', camera_b_distance2(15, 5))
print('f=15mm, g=15m:', camera_b_distance2(15, 15))
```

*ANSWER*:

- When the object distance increases, the CCD (Charge-Coupled Device, imaging sensor) moves closer to the focal length (15 mm). 
- The required adjustment becomes smaller and smaller (changes in the CCD position become smaller and smaller). 
- In the limit of very large object distance, the CCD position converges to the focal length.

```python
G = 1.8
G_mm = G * 1000 # in mm
g = 5
g_mm = g * 1000
f_mm = 5
```

1. A focused image of Thomas is formed inside the camera. At which distance from the lens?

```python
b = camera_b_distance(f_mm, g)
print('f=5mm, g=5m -> b =', b)
```

2. How tall (in mm) will Thomas be on the CCD-chip?

```python
# b/B = g/G -> B = (bG)/g
B = (b * G_mm) / g_mm
print('B =', B)
```
